## Q1

```python
mean = 100
sd = 15
```

```python
def find_mean_and_sd(sample_size):
    num_trials = 1000
    means = np.zeros(num_trials)
    std_devs = np.zeros(num_trials)
    for trial in range(num_trials):
        random_numbers = np.random.normal(mean, sd, sample_size)

        # Calculate mean and standard deviation of the sample
        mean_sample = np.mean(random_numbers)
        std_dev_sample = np.std(random_numbers)

        means[trial] = mean_sample
        std_devs[trial] = std_dev_sample

    return means, std_devs
```

### For 10 numbers

```python
sample_size = 10
random_numbers = np.random.normal(mean, sd, sample_size)
mean_sample = np.mean(random_numbers)
sd_sample = np.std(random_numbers)
print("10 Sampled Numbers:", random_numbers)
print("Mean of the Sample:", mean_sample)
print("Standard Deviation of the Sample:", sd_sample)
```

```python
means_10, std_devs_10 = find_mean_and_sd(10)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.hist(means_10, bins=30, edgecolor='black', alpha=0.7)
plt.title('Frequency Distribution of Means')
plt.xlabel('Mean Value')
plt.ylabel('Frequency')
plt.grid(True)


# Plot the frequency distribution of standard deviations
plt.subplot(1, 2, 2)
plt.hist(std_devs_10, bins=30, edgecolor='black', alpha=0.7)
plt.title('Frequency Distribution of Standard Deviations')
plt.xlabel('Standard Deviation Value')
plt.ylabel('Frequency')
plt.grid(True)

# Show the plots
plt.tight_layout()
plt.show()
```

1. Difference from Question 1 with Respect to the Central Limit Theorem (CLT):
+ With the Beta distribution, the sampling distribution of the mean is not approaching normality as quickly as observed with the normal distribution in Question 2.
+ As the sample size increases, the distribution of sample means tends to become more normal, adhering to the CLT. However, due to the Beta distribution's shape, it will require relatively larger sample sizes to exhibit this behavior compared to the normal distribution case.

2. Sample Size Requirement for Accuracy in Estimating Population Mean:
+ The Beta distribution's complex shape needs a larger sample size to estimate mean more accurately.
+ Larger sample sizes provide more information about the population, leading to more precise estimates

### For 30 points
